#### Carga de archivo CSV, asigna 1 a los que son Z10

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np

siniestros_original = pd.read_csv('Salud2012.csv')
siniestros = pd.DataFrame()

siniestros['EDAD'] = siniestros_original['EDAD']
siniestros['SEXO'] = siniestros_original['SEXO']
siniestros['DIAGNOST1'] = siniestros_original['DIAGNOST1']
siniestros['LUG_RESIDE'] = siniestros_original['LUG_RESIDE']

consulta = siniestros[siniestros['DIAGNOST1'] == 'Z10']
consulta['CONSULTA'] = 1
no_consulta = siniestros[siniestros['DIAGNOST1'] != 'Z10']
no_consulta['CONSULTA'] = 0
siniestros = consulta.append(no_consulta)
siniestros.drop('DIAGNOST1', axis=1, inplace=True)

print siniestros.head()
print siniestros.tail()
```

#### Cambia sexo por variable booleana (1 = hombre)

```python
siniestros_m = siniestros[siniestros['SEXO']=='M']
siniestros_m['MASCULINO'] = 1
siniestros_f = siniestros[siniestros['SEXO']=='F']
siniestros_f['MASCULINO'] = 0
siniestros = pd.concat([siniestros_m, siniestros_f])
siniestros['SEXO'] = siniestros['MASCULINO']
siniestros.drop('MASCULINO', axis=1, inplace=True)

print siniestros.head()
print siniestros.tail()
```

#### Asigna grupo de edad

```python
def asigna_gpo(siniestros):
    edad = siniestros['EDAD']
    if(edad <= 10):
        siniestros['GPO_EDAD'] = 1
    elif(edad > 10 and edad <= 20):
        siniestros['GPO_EDAD'] = 2
    elif(edad > 20 and edad <= 30):
        siniestros['GPO_EDAD'] = 3
    elif(edad > 30 and edad <= 40):
        siniestros['GPO_EDAD'] = 4
    elif(edad > 40 and edad <= 50):
        siniestros['GPO_EDAD'] = 5
    elif(edad > 50 and edad <= 60):
        siniestros['GPO_EDAD'] = 6
    else:
        siniestros['GPO_EDAD'] = 7
    return siniestros
siniestros = siniestros.apply(asigna_gpo, axis=1)

siniestros['EDAD'] = siniestros['GPO_EDAD']
siniestros.drop('GPO_EDAD', axis=1, inplace=True)

print siniestros.head()
print siniestros.tail()
```

#### Asigna región

```python
def asigna_region(siniestros):
    estado = siniestros['LUG_RESIDE']
    region_1 = [5, 19, 28]
    region_2 = [2, 3, 8, 10, 25, 26]
    region_3 = [6, 18, 16, 14]
    region_4 = [13, 21, 29, 30]
    region_5 = [1, 11, 22, 24, 32]
    region_6 = [9, 15, 17]
    region_7 = [4, 23, 27, 31]
    region_8 = [7, 12, 20]
    if estado in region_1:
        siniestros['REGION'] = 1
    elif estado in region_2:
        siniestros['REGION'] = 2
    elif estado in region_3:
        siniestros['REGION'] = 3
    elif estado in region_4:
        siniestros['REGION'] = 4
    elif estado in region_5:
        siniestros['REGION'] = 5
    elif estado in region_6:
        siniestros['REGION'] = 6
    elif estado in region_7:
        siniestros['REGION'] = 7
    else:
        siniestros['REGION'] = 8
    return siniestros
siniestros = siniestros.apply(asigna_region, axis=1)

siniestros['LUG_RESIDE'] = siniestros['REGION']
siniestros.drop('REGION', axis=1, inplace=True)

print siniestros.head()
print siniestros.tail()
```

#### Árbol con edad y sexo

```python
from sklearn import tree
from pandas import read_table, DataFrame

X = siniestros.ix[:, 'EDAD':'SEXO']
Y = siniestros['CONSULTA']

dtree = tree.DecisionTreeClassifier(criterion = "entropy", min_samples_leaf = 500)
dtree = dtree.fit(X, Y)
print DataFrame(dtree.feature_importances_, columns = ["Imp"], index = X.columns).sort(['Imp'], ascending = False)
dotfile = open("edad_sexo.dot", 'w')
dotfile = tree.export_graphviz(dtree, out_file = dotfile, feature_names = X.columns)
dotfile.close()
!dot -Tpng edad_sexo.dot -o edad_sexo.png

from IPython.core.display import Image 
Image(filename='edad_sexo.png')
```

#### Árbol con edad, sexo y región

```python
X = siniestros.ix[:, 'EDAD':'LUG_RESIDE']
Y = siniestros['CONSULTA']

dtree = tree.DecisionTreeClassifier(criterion = "entropy", min_samples_leaf = 500)
dtree = dtree.fit(X, Y)
print DataFrame(dtree.feature_importances_, columns = ["Imp"], index = X.columns).sort(['Imp'], ascending = False)
dotfile = open("edad_sexo_region.dot", 'w')
dotfile = tree.export_graphviz(dtree, out_file = dotfile, feature_names = X.columns)
dotfile.close()
!dot -Tpng edad_sexo_region.dot -o edad_sexo_region.png

from IPython.core.display import Image 
Image(filename='edad_sexo_region.png')
```
